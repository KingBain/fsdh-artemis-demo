```python
database_name = "fsdh_artemis"

raw_path = "dbfs:/FileStore/fsdh_artemis/raw"
bronze_path = "dbfs:/FileStore/fsdh_artemis/bronze"
silver_path = "dbfs:/FileStore/fsdh_artemis/silver"
gold_path = "dbfs:/FileStore/fsdh_artemis/gold"
```

```python
spark.sql(f"CREATE DATABASE IF NOT EXISTS {database_name}")
print(f"Database ready: {database_name}")
```

```python
dbutils.fs.mkdirs(raw_path)
dbutils.fs.mkdirs(bronze_path)
dbutils.fs.mkdirs(silver_path)
dbutils.fs.mkdirs(gold_path)

print("Folders ready:")
print(f"- {raw_path}")
print(f"- {bronze_path}")
print(f"- {silver_path}")
print(f"- {gold_path}")
```

```python
display(spark.sql("SHOW DATABASES"))
```

```python
print("Raw:")
raw_files = dbutils.fs.ls(raw_path)
print(raw_files)

print("Bronze:")
bronze_files = dbutils.fs.ls(bronze_path)
print(bronze_files)

print("Silver:")
silver_files = dbutils.fs.ls(silver_path)
print(silver_files)

print("Gold:")
gold_files = dbutils.fs.ls(gold_path)
print(gold_files)
```
